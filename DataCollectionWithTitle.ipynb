```python
import requests
from bs4 import BeautifulSoup
urls=[]
titles=[]
```

```python
#This program collects the books and create txt files named Book1, Book2.... without book titles
```

```python
for i in range(601,801):
    url="http://www.gutenberg.org/ebooks/"+str(i)
    r=requests.get(url)
    if r.status_code==404:
        print("Book "+str(i)+" doesn't exist.")
        continue
    html_contents=r.text
    html_soup=BeautifulSoup(html_contents,"html.parser")
    #Get book titles
    t=html_soup.find('h1').text.strip()
    #Get the txt file page url for each book
    url2="http://www.gutenberg.org/files/"+str(i)+"/"
    r2=requests.get(url2)
    html_contents2=r2.text
    html_soup2=BeautifulSoup(html_contents2,"html.parser")
    t2=html_soup2.find_all('a')
    length=len(urls)
    for j in t2:
        if "txt" in j.text.strip():
            urls.append("http://www.gutenberg.org/files/"+str(i)+"/"+str(j.text.strip()))
            break
    if length==len(urls):
        continue
    titles.append(t)
    print("Book "+str(i)+" done")
```

```python
#Replace illegal characters in book titles, these illegal characters will cause an error when creating txt files
import re
for j in titles:
    titles[titles.index(j)]=re.sub('[\/:*?"<>|]','-',j)
```

```python
#Get the content of books and write to txt files
for m in range(580,len(urls)):
    i=urls[m]
    r=requests.get(i)
    r.encoding='utf-8'
    html_contents=r.text.replace('\n','').replace('\r','')
    file=open("Book"+str(m+1)+"_"+str(titles[m])+".txt",'w',encoding='utf-8')
    #file=open("Book"+str(m)+".txt",'w',encoding='utf-8')
    file.write(html_contents)
    file.close()
    print('file'+str(m+1)+' done')
```
